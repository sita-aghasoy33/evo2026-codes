```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
def read_data(path):
    # read Customer_Info sheet from Bank_Churn_Customer
    # return the 'EstimatedSalary' column from data
    column = pd.read_excel(path, sheet_name='Customer_Info')['EstimatedSalary']
    return column
```

```python
def wrangle(series, random_state=False, n=500):
    # Clean the string
    new_col = (series
               .str.replace('€', "")
               .astype(float))

    # Select sample
    if random_state:
      return new_col.sample(n=n, random_state = random_state)
    else:
      return new_col.sample(n=n)
```

```python
def z_test(sample, hyp_mean, pop_std, tail='two'):
    assert tail in ['left', 'right', 'two'], "tail must be 'left', 'right' or 'two'"
    conf_level = 0.95
    alpha = round(1 - conf_level, 2)
    # calculate z-score
    z_score = ((sample.mean() - hyp_mean) /
                    (pop_std / np.sqrt(len(sample))))
    # calculate p-value
    if tail == 'left':
      p_value = stats.norm.cdf(z_score)
    elif tail == 'right':
      p_value = 1 - stats.norm.cdf(z_score)
    elif tail == 'two':
      p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    
    # make decision
    if p_value < alpha:
      result = "Reject the null hypothesis"
    else:
      result = "Fail to reject the null hypothesis"
    return z_score, p_value, result
```
