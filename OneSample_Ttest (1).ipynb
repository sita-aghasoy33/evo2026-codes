```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# max row and columns to display
pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)

# width of DataFrame
pd.set_option("display.width", 1000)

# Float number formatting
pd.options.display.float_format = "{:,.2f}".format

# p-value decimal formatting
np.set_printoptions(suppress=True)
```

### 1) Example sample data

```python
data = np.array([480, 510, 530, 495, 505, 520, 515, 490, 500, 525,
                 510, 498, 507, 492, 535, 488, 506, 499, 512, 520])

# Hypothesized population mean
mu0 = 500
```

### 2) Manual t-statistic

```python
n = len(data)
x_bar = np.mean(data)
s = np.std(data, ddof=1)  # sample std (ddof=1)

t_score = (x_bar - mu0) / (s / np.sqrt(n))
print(t_score)
```

### 3) Two-tailed p-value

```python
df = n - 1
p_value = 2 * (1 - stats.t.cdf(abs(t_score), df=df))

print("One-Sample t-test (Two-tailed)")
print("--------------------------------")
print(f"n = {n}")
print(f"Sample mean = {x_bar:.2f}")
print(f"Sample std  = {s:.2f}")
print(f"t-score     = {t_score:.4f}")
print(f"df          = {df}")
print(f"p-value     = {p_value:.6f}")
```

### Decision

```python
# Decision at alpha=0.05
alpha = 0.05
if p_value < alpha:
    print("✅ Reject H0: mean 500-dən fərqlidir.")
else:
    print("❌ Fail to reject H0: mean 500-dən fərqli deməyə sübut yoxdur.")
```
