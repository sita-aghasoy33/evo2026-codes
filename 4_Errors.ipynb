### Types of Semantic Error

Semantic error koda yanlış təlimat verdiyimiz kodlar

```python
# Məqsəd: ədədin kvadratını hesablamaq
x = 5
result = x * 2   # işləyir amma səhvdir
print(result)    # çıxış: 10 (düzgün nəticə 25 olmalı idi)
```

```python
age = 18

# Məqsəd: 18 və yuxarı yaşlıları "Yetkin" demək
if age > 18:          # Burada səhv şərt yazılıb
    print("Yetkin")
else:
    print("Yetkin deyil")
# çıxış: "Yetkin deyil" (amma 18 yaş yetkindir)
```

### Types of Logical Error

```python
# Məqsəd: 1-dən 10-a qədər ədədlərin cəmi
total = 0
for i in range(10):  # səhv diapazon
    total += i
print(total)   # nəticə: 45 (amma düzgün nəticə 55 olmalı idi)
```

Bəzi yerlərdə semantic və logical error arasında fərq qoyulsa da, bununla bağlımübahisəli mənbə çoxdur. Adətən kod doğru planlanıbsa, amma biz niyyətimizi səhv ifadə etmişiksə, buna semantic deyilir. Yox kod yanlış qurulubsa, buna logical deyilir. Lakin çox zaman, bu sözləri əvəzedici kimi də görə bilərsiniz. Ona görə bununla bağlı çox detala enməmək və semantic deyə ümumi ad vermək olar.

Nəzərdə tutduğumuz xətaları qeyd etdikdən sonra, nəticədən asılı olmayaraq kodun icra olunub bitdiyini göstərən və hər bir halda icra olunan bir hissəyə də ehtiyacımız var. Bu zaman "finally" variantını əlavə edirik. Bu hissə digər try və except hissələrindən hansının icra olunmasına baxmayaraq, sonda hər zaman icra olunur.

```python
# @title try except finally
math_quiz = input("Enter a simple math quiz: ")
try:
  answer = eval(math_quiz)
  print(answer)

except NameError:
  print("The variables that you are using in your quiz are not defined. \
  Please, define them first.")

except TypeError:
  print("This is not a valid operation. Please check the operation again.")

finally:
  print("This is the end of the program.")
```

Amma vaxtında oyanmayanda zəngli saatın verdiyi narahatlıq sizin əlinizdə olan bir narahatlıqdır və sizi gələcək pis xətalardan qoruyur.

```python
def generate_cover_letter(name, work_place, job, experience, education):
  """
  params:
    name (str): full name of user.
    work_place (str): current workplace of user.
    job (str): current position of user.
    experience (int): total experience in years.
    education (dict): major and university name of user.
                      {"major": "...", "school": "..."}

  returns:
    cover_letter (str): generated cover letter.
  """

  intro = f"My name is {name}. "
  job_desc = f"Currently I am working as a {job} for {work_place}. "
  exper_info = f"I have a {experience} years of experience in this sector totally. "
  education = f"I have studied {education["major"]} in {education["school"]}. "
  final = "Looking forward to hearing from you."

  return intro + job_desc + exper_info + education + final
```

Bəzən də xəta çıxır amma açıqlaıyıcı olmadığı üçün istifadəçi narazı qalır. Bu zaman xətanı biz özümüz yaratmalı və izahedici açıqlama verməliyik. Bunun üçün həm "if else"dən, həm də "try catch"dən istifadə edə bilərik.

```python
def write_cov_letter(name, work_place, job, experience, education):
  """
  params:
    name (str): full name of user.
    work_place (str): current workplace of user.
    job (str): current position of user.
    experience (int): total experience in years.
    education (dict): major and university name of user.
                      {"major": "...", "school": "..."}
  """

  if type(name) != str:
    raise TypeError("Name should be a string!")
  elif name.find(" ") == -1:
    raise ValueError("Name should have first and last name!")

  if type(experience) != int:
    raise TypeError("Experience should be an integer!")
  elif experience < 0:
    raise ValueError("Experience should be a positive number!")

  if type(education) != dict:
    raise TypeError("Education should be a dictionary!")
  elif education.keys() != {"major", "school"}:
    raise ValueError("Education should have 'major' and 'school' keys!")

  intro = f"My name is {name}. "
  job_desc = f"Currently I am working as a {job} for {work_place}. "
  exper_info = f"I have a {experience} years of experience in this sector totally. "
  education = f"I have studied {education["major"]} in {education["school"]}. "
  final = "Looking forward to hearing from you."

  return intro + job_desc + exper_info + education + final
```

Özümüz error yarada bilərik. Class-lara keçid. Pythonda bəhs etdiyimiz primitive və non-primitive data tipləri kimi hər bir şeyin tipi olur. Biz özümüz də bu tipləri formalaşdıra və o tipdə obyektlər formalaşdıra bilərik ki, bu da OOP sahəsinə aid mövzudur. Bunun ən məşhur nümunəsi də error növləri yaratmaqdır. Bu bəzən istifadəçi üçün kodu daha oxunaqlı etməkdə kömək edir, ya da o an fərqli xətaya ehtiyacımız olur.

```python
class AgeError(Exception):
    pass

def check_age(age):
    if age < 18:
        raise AgeError("Yaş çox azdır!")

try:
    check_age(15)
except AgeError as e:
    print("Xəta:", e)
```
