## Two-Tailed Proportion Z-test (One Sample)

### Define the Hypothesis

- H0
- H1

```python
# Given data
x = 62      # successes
n = 100     # sample size
p0 = 0.50   # hypothesized proportion (H0)
```

```python
# 1) Sample proportion
p_hat = x / n
print(p_hat)
```

```python
# 2) Standard error under H0
SE = np.sqrt(p0 * (1 - p0) / n)
print(SE)
```

```python
# 3) Z-score
z = (p_hat - p0) / SE
print(z)
```

```python
# 4) Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))
print(p_value)
```

```python
# 4) Right-tailed p-value
p_value = 1 - stats.norm.cdf(z)
print(p_value)
```

```python
# 4) Left-tailed p-value
p_value = stats.norm.cdf(z)
print(p_value)
```

```python
print("Sample proportion (p̂):", p_hat)
print("Standard Error (SE):", SE)
print("Z-score:", z)
print("P-value (two-tailed):", p_value)
```

```python
# 5) Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0 ✅ (significant difference)")
else:
    print("Fail to reject H0 ❌ (no significant difference)")
```
