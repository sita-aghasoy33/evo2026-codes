```python
# CREATE CONNECTION WITH ANOTHER METHOD (SQL ALCHEMY)
user_name = "system"
password = "****"
host = "127.0.0.1" # localhost
port = "1521/"
service_name = "ORCLPDB" # service name for the databse we are accessing

engine = create_engine(
    "oracle+oracledb://"+user_name+":"+password+"@"+host+":"+port+"?service_name="+service_name
)

df = pd.read_sql("SELECT * FROM CUSTOMER_INFO", engine)
df
```
