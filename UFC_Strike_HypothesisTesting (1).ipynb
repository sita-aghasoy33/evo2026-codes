## Hypothesis Testing

### Two-tailed z-test

- H0 ---> sample mean = hypothetical population mean.
- H1 ---> sample mean <> hypothetical population mean.

```python
alpha = 0.05
```

```python
population_mean = 41
population_std = 14.8
sample_mean = float(strike_acc_sample.mean())
sample_size = strike_acc_sample.shape[0]
print(sample_mean)
```

```python
strike_acc.std()
```

```python
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
z_score
```

```python
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
print(p_value)
```

```python
# @title 5. Final Decision
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.5f}")

if p_value < alpha:
    print("Reject H0 → Statistically significant difference.")
else:
    print("Fail to reject H0 → No statistically significant difference.")
```

### Right-tailed z-test

- H0 ---> sample mean is not greater than population mean (less and equal).
- H1 ---> sample mean is greater than population mean.

```python
alpha = 0.05
```

```python
population_mean = 40
population_std = 14.8
sample_mean = float(strike_acc_sample.mean())
sample_size = strike_acc_sample.shape[0]
print(sample_mean)
```

```python
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
z_score
```

```python
# @title 4. Find p-value
p_value = 1 - stats.norm.cdf(z_score)
print(p_value)
```

```python
# @title 5. Final Decision
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.5f}")

if p_value < alpha:
    print("Reject H0 → The sample mean is statistically significantly higher than the population mean.")
else:
    print("Fail to reject H0 → The sample mean is not statistically significantly higher than the population mean.")
```

### Left-tailed z-test

- H0 ---> sample mean is **not less** than population mean (greater and equal).
- H1 ---> sample mean is **less** than population mean.

```python
population_mean = 41
population_std = 14.8
sample_mean = float(strike_acc_sample.mean())
sample_size = strike_acc_sample.shape[0]
print(sample_mean)
```

```python
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
z_score
```

```python
# @title 4. Find p-value
p_value = stats.norm.cdf(z_score)
print(p_value)
```

```python
# @title 5. Final Decision
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.5f}")

if p_value < alpha:
    print("Reject H0 → The sample mean is statistically significantly lower than the population mean.")
else:
    print("Fail to reject H0 → The sample mean is not statistically significantly lower than the population mean.")
```
