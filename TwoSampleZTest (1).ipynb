```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# max row and columns to display
pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)

# width of DataFrame
pd.set_option("display.width", 1000)

# Float number formatting
pd.options.display.float_format = "{:,.2f}".format

# p-value decimal formatting
np.set_printoptions(suppress=True)
```

### 1) Sample data (example)

```python
np.random.seed(23)

# Group 1 and Group 2 random samples
# (these are just example samples)
n1 = 50
n2 = 60

group1 = np.random.normal(loc=520, scale=30, size=n1) # from 1st population
group2 = np.random.normal(loc=500, scale=30, size=n2) # from 2nd population

print("group1 average salary:", group1.mean())
print("group2 average salary:", group2.mean())

print("group1 std:", group1.std())
print("group2 std:", group2.std())
```

### 2) Known population std devs (Z-test requires this)

```python
sigma1 = 30   # known population std for group1
sigma2 = 30   # known population std for group2
```

### 3) Compute sample means

```python
x1_bar = np.mean(group1)
x2_bar = np.mean(group2)
print("1st sample mean:", x1_bar)
print("2nd sample mean:", x2_bar)
```

### 4) Z-score

```python
standard_error = np.sqrt((sigma1**2 / n1) + (sigma2**2 / n2))
z_score = (x1_bar - x2_bar) / standard_error
print(z_score)
```

### 5) Two-tailed p-value

```python
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
print(p_value)
```

### 6) Decision (alpha=0.05)

```python
alpha = 0.05

print("Two-Sample Z-Test (Two-Tailed)")
print("--------------------------------")
print(f"n1 = {n1}, mean1 = {x1_bar:.2f}, sigma1 = {sigma1}")
print(f"n2 = {n2}, mean2 = {x2_bar:.2f}, sigma2 = {sigma2}")
print(f"Z-score = {z_score:.4f}")
print(f"P-value  = {p_value:.6f}")

if p_value < alpha:
    print("✅ Reject H0: ortalamalar arasında statistik fərq var.")
else:
    print("❌ Fail to reject H0: ortalamalar arasında fərq üçün sübut yoxdur.")
```
